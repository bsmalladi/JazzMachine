```python
d = mus.duration.Duration (4.0 * choMeas)
C7 = mus.chord.Chord ('C4 E4 G4 B-4', duration = d)
F7 = mus.chord.Chord ('F4 A4 C5 E-5', duration = d)
G7 = mus.chord.Chord ('G4 B4 D5 F5', duration = d)
C7.volume = mus.volume.Volume (velocity = 40)
F7.volume = mus.volume.Volume (velocity = 40)
G7.volume = mus.volume.Volume (velocity = 40)

def chooseArpNotes (arp, arpTot):
    ivs = [-2, -1, 0, 1, 2]
    pbIvs = [.1, .3, .2, .3, .1]
    curNote = int (round (len (arp) / 2, 0))
    out = arpTot
    for i in range (choMeas * 2):
        prob = np.random.multinomial (1, pbIvs)
        ind = np.where (prob) [0][0]
        if curNote + ivs [ind] >= len (arp) or curNote + ivs [ind] < 0:
            curNote = curNote - ivs [ind]
        else:
            curNote = curNote + ivs [ind]
        out.append (arp [curNote])                   #returns array of note vars
    #print(out)
    return out


def choosePassingNotes (arpMel, scaleRegArr, scaleFunkArr):   #takes entire arp melody and adds passing notes
    out = arpMel
    for k in range(len(scaleRegArr)):
        for i in range (choMeas * 2):
            print(" i = " + str(i) + " k = " + str(k))
            print(out)
            ivs = [0, 1, 2]
            pbIvs = [.25, .55, .2]
            prob = np.random.multinomial (1, pbIvs)
            ind = np.where (prob) [0][0]
        
            if ivs[ind] == 0:
                out.insert(k*2 + i, note_rest)
            else:
                if ivs[ind] == 1:
                    out.insert(k + i*2), scaleRegArr[k][findNote (arpMel [k + i*2], arpMel [k + i*2 + 1], scaleRegArr[k])])
                else:
                    if (scaleFunkArr[k].index(arpMel[k + i*2]) == scaleFunkArr[k].index(arpMel[k + i*2 +1])) or (abs(scaleFunkArr[k].index(arpMel[k + i*2]) - scaleFunkArr[k].index(arpMel[k + i*2 +1])) > 3):
                        out.insert(k + i*2, findNote (arpMel [k + i*2], arpMel [k + i*2 + 1], scaleRegArr[k]))
                    else:
                        out.insert(k + i*2, makeTrip(arpMel[k + i*2], arpMel[k + i*2 + 1], scaleFunkArr[k]))
    out.append (arpMel [-1])
    return out


def makeTrip (note1, note2, funk):
    ind1 = funk.index(note1)
    ind2 = funk.index(note2)
    if ind1 < ind2:
        note3 = funk[ind1 + 1]
        note4 = funk[ind1 + 2]
        triplet = "trip{" + note1 + "4 " + note3 + " " + note4 + "}"
        return triplet
    else:
        note3 = funk[ind1 - 1]
        note4 = funk[ind1 - 2]
        triplet = "trip{" + note1 + "4 " + note3 + " " + note4 + "}"
        return triplet
    
    
def findNote (note1, note2, scale):      #only for quarter notes
    #if abs (ind1 - ind2) == 2:
    ind1 = scale.index(note1)
    ind2 = scale.index(note2)
    if ind1 != ind2:
        return int (round ((ind1 + ind2) / 2, 0))
    else:
        ivs = [-1, 1]
        pbIvs = [.5, .5]
        prob = np.random.multinomial (1, pbIvs)
        ind = np.where (prob) [0][0]
        ud = ivs [ind]
        if ind1 + ud >= len (Cmaj) or ind1 + ud < 0:
            ud = ud * -1
        return ind1 + ud


def outString (indMel):
    out = []
    for i in range (len (indMel)):
        if indMel [i] is None:
            out.append ("r")
        else:
            out.append (Cmaj [indMel [i]])
    return out


'''def measMelody (measCho, arp, scale, funky):
    mel = chooseArpNotes (arp)
    mel2 = choosePassingNotes (mel, scale, funky)
    #print(mel2)
    return mel2
'''


def totArp (arps):
    arpTot = []
    for i in range(len(arps)):
        arpTot = chooseArpNotes(arps[i], arpTot)
    return arpTot



def melTot (arps, chords, funkArr, regArr):
    totMel = []
    totMel = choosePassingNotes(totArp(arps), regArr, funkArr)
    print (totMel)
    tnMelody = 'tinyNotation: ' + " ".join(totMel)
    out = mus.converter.parse(tnMelody)
    for i in range (len (chords)):
        cx = copy.deepcopy (chords [i])
        out.insert (i * 4 * choMeas, cx)
    return out
```
